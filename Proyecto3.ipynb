(Eliminación o rellenado de Valores Faltantes)**

En el tratamiento de los datos faltantes, realizamos dos imputaciones para la primera ponemos el valor de una media general en los faltantes y para la segunda ponemos el valor de ua media condicionada por grupos en los faltaste haciendo esto el tratamiento o manipulación de datos faltantes de forma más robusta.

Segunda imputación : Inyectamos una media condicionada a los datos nulos a traves de las tablas pivot,debido  a que este metodo nos otorga un mayor grado de robustes que la imputación 1. Para realizar este proceso es necesario crear grupos etarios para poder verlo más detallado.

````python
data
````

````python
data["age_2"] = pd.cut(data["age"],4)
````

Cremos una tabla pivot  que muestre la media del salario a partir del grupo etario y el sexo 


````python
pivot=data.pivot_table(index=["age_2","gender"],values="wage",aggfunc="mean")
pivot
````

Creamos una función que nos permite verificar los datos faltantes y hacer el reemplazo.

````python
def funcion(row):
  wage=row["wage"]
  age=row["age_2"]
  gender=row["gender"]
  if pd.isna(wage):
    return pivot["wage"][age][gender]
  return wage
````

Realizamos el llenado de la media condicional sobre nuestro base de dato original

````python
data["wage"]=data.apply(funcion,axis=1)
````

````python
data
````

Al crear la función verificamos que ya no existan datos faltantes.

````python
data.isna().sum()
````

**Outliders**
 Primero identificamos el primer cuartil, el tercer cuartil,el rango intercuartilico, y el limite superior e inferior de la distribucion de los datos de la variable wage 

````python
variable_wage=data.loc[::1,"wage"]
````

````python
variable_wage
````

````python
variable_wage.describe()
````

````python
Q_1=variable_wage.quantile(0.25)
````

````python
Q_1
````

````python
Q_3=variable_wage.quantile(0.75)
````

````python
Q_3
````

````python
rango_intercu=Q_3-Q_1
````

````python
rango_intercu
````

````python
limite_sup=(rango_intercu*1.5)+Q_3
limite_sup
````

````python
limite_inf=Q_1-(1.5*rango_intercu)
limite_inf
````

Se determinan los datos que mayores al limite superior y se toma su posiion dentro del dataset 

````python
outlier=variable_wage[(variable_wage>limite_sup)]
````

````python
indice_out=outlier.index
````

````python
indice_out
````

Se eliminan los outliers

````python
no_outliers=variable_wage[~variable_wage.index.isin(indice_out)]
````

````python
no_outliers.describe()
````

````python
outliers_2=no_outliers[no_outliers<min([limite_inf])]
outliers_2
````
